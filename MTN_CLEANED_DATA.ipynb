```python
#Coverting Date column to Date Data Type

MTN_df['Date'] =pd.to_datetime(MTN_df['Date'])
```

```python
#setting date as index
MTN_df.set_index('Date', inplace=True)
```

```python
###Removing spaces(Trim) in columns

MTN_df.columns = MTN_df.columns.str.strip()
```

```python
MTN_df
```

```python
MTN_df= MTN_df.drop(columns=['Change %'])
```

```python
# Cleaning and converting the values in the Volume column


MTN_df['Vol.'] = MTN_df['Vol.'].astype(str).str.upper()  ### Convert Vol. column to string and uppercase everything


clean_volume = []                             # Create an empty list to hold the cleaned numbers


for val in MTN_df['Vol.']:                 #Loop through the values to clean them
    if 'K' in val:
        num = float(val.replace('K', ''))
        clean_volume.append(num * 1000)
    elif 'M' in val:
        num = float(val.replace('M', ''))
        clean_volume.append(num * 1000000)
    else:
        clean_volume.append(float(val))


MTN_df['Vol.'] = clean_volume                    # Assign the cleaned values back to the Vol. column
```
